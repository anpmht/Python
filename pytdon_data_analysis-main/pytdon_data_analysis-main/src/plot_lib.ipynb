```python
import pandas as pd
import matplotlib.pyplot as plt

# class to process and plot data
class dataPlot:
    
    # init function to initialize the obect
    def __init__(self, file_Name):
        
        self.file_Name = file_Name
        self.df = pd.read_csv(file_Name)
        self.locations = ["empty", "BL", "BR", "SL", "SR", "LY", "LZ", "LX", "time"]
        print("strain gauge location\t:corrosponding amplifer")
        for i in range(len(self.df.columns)):
            print(self.df.columns[i],"\t: ",self.locations[i])
        
        print("\ncall the plot function as following line")
        print("c1.plot(col_name, from, to, color, xlable, ylable)")
        
    #plot single line on single chart
    def plot_seperate(self,colum_No,RangeFrom,RangeTo,color,xlabel,ylabel):
        self.RangeFrom = RangeFrom
        self.RangeTo = RangeTo
        self.colum_No = colum_No
        self.color = color
        self.xlabel = xlabel
        self.ylabel = ylabel
        Title = self.colum_No

        for i in range(len(self.df.columns)):
            if self.locations[i] == self.colum_No:
                column = self.df.columns[i]
                    
        plt.figure(figsize=(20, 7))
        self.df[column][self.RangeFrom:self.RangeTo].plot(color = color)
        plt.xlabel(self.xlabel)
        plt.ylabel(self.ylabel)
        plt.title(Title)

    # plot multple lines on the same chart
    def plot_together(self,plot_columns,RangeFrom,RangeTo,xlabel,ylabel):
        self.RangeFrom = RangeFrom
        self.RangeTo = RangeTo
        self.plot_columns = plot_columns
        self.xlabel = xlabel
        self.ylabel = ylabel        
        c = 0
        colors = ["red", "green", "blue", "darkred", "deeppink", "yellow", "maroon", "orange"]
        plt.figure(figsize=(20, 8))
        Title = str(plot_columns)
        plt.title(Title)
        for i in self.plot_columns:
            for j in range(len(self.df.columns)):
                if self.locations[j] == i:
                    column = self.df.columns[j]
            color = colors[c]
            self.df[column][self.RangeFrom:self.RangeTo].plot(color = color)
            plt.xlabel(self.xlabel)
            plt.ylabel(self.ylabel)
            c += 1
    
    
    
    # normalize data to a certain range for easier analysis
    def normalizedata(self):
        pass
    
    # fix the initial offsets
    def fixOffset(self):
        pass
```
